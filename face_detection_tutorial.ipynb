# 1ë‹¨ê³„: í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜í•˜ê¸° ğŸ“¦

ì•„ë˜ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ë©´ ë‹¤ìŒê³¼ ê°™ì€ ë„êµ¬ë“¤ì´ ì„¤ì¹˜ë©ë‹ˆë‹¤:
- **opencv-python**: ì¹´ë©”ë¼ì™€ ì´ë¯¸ì§€ë¥¼ ë‹¤ë£¨ëŠ” ë„êµ¬
- **ultralytics**: YOLO AI ëª¨ë¸ì„ ì‚¬ìš©í•˜ëŠ” ë„êµ¬
- **numpy**: ìˆ˜í•™ ê³„ì‚°ì„ ë„ì™€ì£¼ëŠ” ë„êµ¬
- **dlib**: ì–¼êµ´ ë¶„ì„ì„ ë„ì™€ì£¼ëŠ” ë„êµ¬

â° **ì„¤ì¹˜ ì‹œê°„**: ì•½ 2-3ë¶„ ì •ë„ ê±¸ë¦´ ìˆ˜ ìˆì–´ìš”. ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”!

````python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë“¤ì„ ì„¤ì¹˜í•©ë‹ˆë‹¤
# ì´ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ë©´ ìë™ìœ¼ë¡œ ì„¤ì¹˜ê°€ ì§„í–‰ë©ë‹ˆë‹¤

print("ğŸ“¦ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
print("â˜• ì ê¹ ì»¤í”¼ë¼ë„ ë§ˆì‹œë©° ê¸°ë‹¤ë ¤ì£¼ì„¸ìš”!")

import sys
import subprocess
import os

def install_packages():
    """
    í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë“¤ì„ ì„¤ì¹˜í•˜ëŠ” í•¨ìˆ˜
    macOSì˜ ì™¸ë¶€ ê´€ë¦¬ í™˜ê²½ ë¬¸ì œë¥¼ í•´ê²°í•©ë‹ˆë‹¤
    """
    packages = ['opencv-python', 'ultralytics', 'numpy']
    
    for package in packages:
        try:
            print(f"ğŸ”§ {package} ì„¤ì¹˜ ì¤‘...")
            
            # ì²« ë²ˆì§¸ ì‹œë„: ì¼ë°˜ ì„¤ì¹˜
            try:
                result = subprocess.run(
                    [sys.executable, '-m', 'pip', 'install', package], 
                    capture_output=True, text=True, check=True
                )
                print(f"âœ… {package} ì„¤ì¹˜ ì™„ë£Œ!")
                continue
            except subprocess.CalledProcessError as e:
                if "externally-managed-environment" in str(e.stderr):
                    print(f"   âš ï¸  ì™¸ë¶€ ê´€ë¦¬ í™˜ê²½ ë¬¸ì œ ê°ì§€. ë‹¤ë¥¸ ë°©ë²•ìœ¼ë¡œ ì‹œë„í•©ë‹ˆë‹¤...")
                    
                    # ë‘ ë²ˆì§¸ ì‹œë„: --break-system-packages í”Œë˜ê·¸ ì‚¬ìš©
                    try:
                        result = subprocess.run(
                            [sys.executable, '-m', 'pip', 'install', '--break-system-packages', package], 
                            capture_output=True, text=True, check=True
                        )
                        print(f"âœ… {package} ì„¤ì¹˜ ì™„ë£Œ! (ì‹œìŠ¤í…œ íŒ¨í‚¤ì§€ ìš°íšŒ ë°©ì‹)")
                        continue
                    except subprocess.CalledProcessError as e2:
                        print(f"   âŒ ì‹œìŠ¤í…œ íŒ¨í‚¤ì§€ ìš°íšŒ ë°©ì‹ë„ ì‹¤íŒ¨: {e2.stderr}")
                        
                        # ì„¸ ë²ˆì§¸ ì‹œë„: --user í”Œë˜ê·¸ ì‚¬ìš©
                        try:
                            result = subprocess.run(
                                [sys.executable, '-m', 'pip', 'install', '--user', package], 
                                capture_output=True, text=True, check=True
                            )
                            print(f"âœ… {package} ì„¤ì¹˜ ì™„ë£Œ! (ì‚¬ìš©ì ì„¤ì¹˜ ë°©ì‹)")
                            continue
                        except subprocess.CalledProcessError as e3:
                            print(f"   âŒ ì‚¬ìš©ì ì„¤ì¹˜ ë°©ì‹ë„ ì‹¤íŒ¨: {e3.stderr}")
                            print(f"\nğŸ”§ {package} ìˆ˜ë™ ì„¤ì¹˜ ë°©ë²•:")
                            print(f"   í„°ë¯¸ë„ì—ì„œ ë‹¤ìŒ ëª…ë ¹ì–´ë¥¼ ì‹¤í–‰í•´ë³´ì„¸ìš”:")
                            print(f"   pip3 install --break-system-packages {package}")
                            print(f"   ë˜ëŠ”: brew install {package}")
                            return False
                else:
                    print(f"âŒ {package} ì„¤ì¹˜ ì‹¤íŒ¨: {e}")
                    print(f"   ì˜¤ë¥˜ ë‚´ìš©: {e.stderr}")
                    return False
                    
        except Exception as e:
            print(f"âŒ {package} ì„¤ì¹˜ ì¤‘ ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜: {e}")
            return False
    
    return True

# ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ ì‹¤í–‰
print("ğŸ’¡ macOSì—ì„œ ì™¸ë¶€ ê´€ë¦¬ í™˜ê²½ ë¬¸ì œê°€ ìˆì„ ìˆ˜ ìˆì–´ìš”. ì—¬ëŸ¬ ë°©ë²•ìœ¼ë¡œ ì‹œë„í•´ë´…ë‹ˆë‹¤!")
success = install_packages()

if success:
    print("\nğŸ‰ ëª¨ë“  ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ê°€ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
    print("âœ¨ OpenCV, YOLO, NumPyê°€ ì¤€ë¹„ë˜ì—ˆì–´ìš”!")
    print("ğŸ“ ì´ì œ dlib ì„¤ì¹˜ë¥¼ ì§„í–‰í•´ë´…ì‹œë‹¤.")
else:
    print("\nâš ï¸  ì„¤ì¹˜ ì¤‘ ë¬¸ì œê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤.")
    print("ğŸ”§ ë‹¤ìŒ í•´ê²° ë°©ë²•ë“¤ì„ ì‹œë„í•´ë³´ì„¸ìš”:")
    print("\nğŸ“‹ í•´ê²° ë°©ë²• 1: í„°ë¯¸ë„ì—ì„œ ì§ì ‘ ì„¤ì¹˜")
    print("   pip3 install --break-system-packages opencv-python ultralytics numpy")
    print("\nğŸ“‹ í•´ê²° ë°©ë²• 2: Homebrewë¡œ ì„¤ì¹˜")
    print("   brew install opencv python-numpy")
    print("   pip3 install --break-system-packages ultralytics")
    print("\nğŸ“‹ í•´ê²° ë°©ë²• 3: ê°€ìƒí™˜ê²½ ì‚¬ìš©")
    print("   python3 -m venv face_emotion_env")
    print("   source face_emotion_env/bin/activate")
    print("   pip install opencv-python ultralytics numpy")
    print("\nğŸ’­ ì–´ë ¤ìš°ë©´ ì„ ìƒë‹˜ì´ë‚˜ ì¹œêµ¬ì—ê²Œ ë„ì›€ì„ ìš”ì²­í•˜ì„¸ìš”!")
````

# 4ë‹¨ê³„: ê°ì • ë¶„ì„ í•¨ìˆ˜ ë§Œë“¤ê¸° ğŸ˜ŠğŸ˜¢ğŸ˜ ğŸ˜²

## ì–´ë–»ê²Œ ê°ì •ì„ ì•Œì•„ë‚¼ê¹Œìš”?

ìš°ë¦¬ í•¨ìˆ˜ëŠ” ì–¼êµ´ì˜ **68ê°œ ì¤‘ìš” í¬ì¸íŠ¸**ë¥¼ ë¶„ì„í•©ë‹ˆë‹¤:

### 1. ì… ëª¨ì–‘ ë¶„ì„ ğŸ‘„
- **ì›ƒìŒ**: ì…ê¼¬ë¦¬ê°€ ì˜¬ë¼ê°€ê³  ì…ì´ ì˜†ìœ¼ë¡œ ê¸¸ì–´ì§
- **ìŠ¬í””**: ì…ê¼¬ë¦¬ê°€ ì•„ë˜ë¡œ ë‚´ë ¤ê°

### 2. ëˆˆ í¬ê¸° ë¶„ì„ ğŸ‘€
- **ë†€ëŒ**: ëˆˆì´ í‰ì†Œë³´ë‹¤ í¬ê²Œ ëœ¨ì„
- **í‰ìƒì‹œ**: ì ë‹¹í•œ í¬ê¸°

### 3. ëˆˆì¹ ìœ„ì¹˜ ë¶„ì„ ğŸ¤¨
- **í™”ë‚¨**: ëˆˆì¹ì´ ë‚´ë ¤ì™€ì„œ ì°¡ê·¸ë¦° ëª¨ì–‘
- **ë†€ëŒ**: ëˆˆì¹ì´ ì˜¬ë¼ê°

### ê°ì§€í•  ìˆ˜ ìˆëŠ” ê°ì •ë“¤:
- ğŸ˜Š **Happy** (í–‰ë³µ) - ì´ˆë¡ìƒ‰ìœ¼ë¡œ í‘œì‹œ
- ğŸ˜¢ **Sad** (ìŠ¬í””) - ë¹¨ê°„ìƒ‰ìœ¼ë¡œ í‘œì‹œ  
- ğŸ˜  **Angry** (í™”ë‚¨) - ì§„í•œ ë¹¨ê°„ìƒ‰ìœ¼ë¡œ í‘œì‹œ
- ğŸ˜² **Surprised** (ë†€ëŒ) - ë…¸ë€ìƒ‰ìœ¼ë¡œ í‘œì‹œ
- ğŸ˜ **Neutral** (í‰ìƒì‹œ) - í°ìƒ‰ìœ¼ë¡œ í‘œì‹œ

````python
def analyze_facial_expression(landmarks):
    """
    ì–¼êµ´ ëœë“œë§ˆí¬ë¥¼ ë¶„ì„í•˜ì—¬ ê°ì •ì„ íŒë‹¨í•˜ëŠ” í•¨ìˆ˜
    
    ì…ë ¥: landmarks (ì–¼êµ´ì˜ 68ê°œ ì¤‘ìš” í¬ì¸íŠ¸)
    ì¶œë ¥: (ê°ì •ì´ë¦„, ìƒ‰ìƒ, ì‹ ë¢°ë„)
    """
    
    # ğŸ—£ï¸ ì… ê´€ë ¨ í¬ì¸íŠ¸ë“¤ì„ ê°€ì ¸ì˜µë‹ˆë‹¤
    mouth_left = landmarks.part(48)    # ì… ì™¼ìª½ ë
    mouth_right = landmarks.part(54)   # ì… ì˜¤ë¥¸ìª½ ë
    mouth_top = landmarks.part(51)     # ì… ìœ„ìª½
    mouth_bottom = landmarks.part(57)  # ì… ì•„ë˜ìª½
    
    # ğŸ‘€ ëˆˆ ê´€ë ¨ í¬ì¸íŠ¸ë“¤ì„ ê°€ì ¸ì˜µë‹ˆë‹¤
    left_eye_top = landmarks.part(37)     # ì™¼ìª½ ëˆˆ ìœ„
    left_eye_bottom = landmarks.part(41)  # ì™¼ìª½ ëˆˆ ì•„ë˜
    right_eye_top = landmarks.part(43)    # ì˜¤ë¥¸ìª½ ëˆˆ ìœ„
    right_eye_bottom = landmarks.part(47) # ì˜¤ë¥¸ìª½ ëˆˆ ì•„ë˜
    
    # ğŸ¤¨ ëˆˆì¹ ê´€ë ¨ í¬ì¸íŠ¸ë“¤ì„ ë¶„ì„í•©ë‹ˆë‹¤
    left_eyebrow = np.mean([landmarks.part(i).y for i in range(17, 22)])   # ì™¼ìª½ ëˆˆì¹ í‰ê·  ë†’ì´
    right_eyebrow = np.mean([landmarks.part(i).y for i in range(22, 27)])  # ì˜¤ë¥¸ìª½ ëˆˆì¹ í‰ê·  ë†’ì´
    
    # ğŸ“Š íŠ¹ì§•ë“¤ì„ ê³„ì‚°í•©ë‹ˆë‹¤
    
    # 1ï¸âƒ£ ì…ì˜ ê°€ë¡œì„¸ë¡œ ë¹„ìœ¨ (ì›ƒì„ ë•ŒëŠ” ì…ì´ ì˜†ìœ¼ë¡œ ê¸¸ì–´ì ¸ìš”)
    mouth_width = abs(mouth_right.x - mouth_left.x)      # ì… ë„ˆë¹„
    mouth_height = abs(mouth_top.y - mouth_bottom.y)     # ì… ë†’ì´
    mouth_ratio = mouth_height / mouth_width if mouth_width > 0 else 0
    
    # 2ï¸âƒ£ ì…ê¼¬ë¦¬ ê¸°ìš¸ê¸° (ì›ƒìœ¼ë©´ ì˜¬ë¼ê°€ê³ , ìŠ¬í”„ë©´ ë‚´ë ¤ê°€ìš”)
    mouth_curve = (mouth_left.y + mouth_right.y) / 2 - mouth_top.y
    
    # 3ï¸âƒ£ ëˆˆì˜ í¬ê¸° (ë†€ë¼ë©´ ëˆˆì´ ì»¤ì ¸ìš”)
    left_eye_height = abs(left_eye_top.y - left_eye_bottom.y)
    right_eye_height = abs(right_eye_top.y - right_eye_bottom.y)
    avg_eye_height = (left_eye_height + right_eye_height) / 2
    
    # 4ï¸âƒ£ ëˆˆì¹ê³¼ ëˆˆ ì‚¬ì´ì˜ ê±°ë¦¬ (í™”ë‚˜ë©´ ëˆˆì¹ì´ ë‚´ë ¤ì™€ìš”)
    avg_eyebrow_height = (left_eyebrow + right_eyebrow) / 2
    eye_center = (left_eye_top.y + right_eye_top.y) / 2
    eyebrow_distance = eye_center - avg_eyebrow_height
    
    # ğŸ¯ ê°ì •ì„ íŒë‹¨í•©ë‹ˆë‹¤!
    emotion = "Neutral"         # ê¸°ë³¸ê°’: í‰ìƒì‹œ
    color = (255, 255, 255)     # ê¸°ë³¸ ìƒ‰ìƒ: í°ìƒ‰
    confidence = 0.5            # ê¸°ë³¸ ì‹ ë¢°ë„: 50%
    
    # ì¡°ê±´ì— ë”°ë¼ ê°ì •ì„ íŒë³„í•©ë‹ˆë‹¤
    if mouth_curve < -5 and mouth_ratio < 0.5:  # ğŸ˜Š ì›ƒìŒ: ì…ê¼¬ë¦¬ ì˜¬ë¼ê°€ê³  ì…ì´ ë„“ì–´ì§
        emotion = "Happy"
        color = (0, 255, 0)  # ì´ˆë¡ìƒ‰
        confidence = min(1.0, abs(mouth_curve) / 10)
        
    elif mouth_curve > 7:  # ğŸ˜¢ ìŠ¬í””: ì…ê¼¬ë¦¬ ë‚´ë ¤ê°
        emotion = "Sad"
        color = (255, 0, 0)  # ë¹¨ê°„ìƒ‰
        confidence = min(1.0, mouth_curve / 10)
        
    elif avg_eye_height > 8 and eyebrow_distance > 15:  # ğŸ˜² ë†€ëŒ: ëˆˆ í¬ê²Œ ëœ¨ê³  ëˆˆì¹ ì˜¬ë¼ê°
        emotion = "Surprised"
        color = (0, 255, 255)  # ë…¸ë€ìƒ‰
        confidence = min(1.0, avg_eye_height / 10)
        
    elif eyebrow_distance < 10:  # ğŸ˜  í™”ë‚¨: ëˆˆì¹ì´ ë‚´ë ¤ì™€ì„œ ì°¡ê·¸ë¦¼
        emotion = "Angry"
        color = (0, 0, 255)  # ì§„í•œ ë¹¨ê°„ìƒ‰
        confidence = min(1.0, (15 - eyebrow_distance) / 10)
    
    return emotion, color, confidence

print("âœ… ê°ì • ë¶„ì„ í•¨ìˆ˜ê°€ ì¤€ë¹„ë˜ì—ˆìŠµë‹ˆë‹¤!")
print("ì´ì œ AI ëª¨ë¸ë“¤ì„ ë¶ˆëŸ¬ì™€ë´…ì‹œë‹¤.")
````

4. ì–¼êµ´ ì£¼ë³€ì— ìƒ‰ê¹”ì´ ìˆëŠ” ì™¸ê³½ì„ ì´ ë‚˜íƒ€ë‚˜ê³  ê°ì •ì´ í‘œì‹œë©ë‹ˆë‹¤

- **ë°©ë²• 3**: í‚¤ë³´ë“œì—ì„œ **'q'** í‚¤ ëˆ„ë¥´ê¸°

## âš ï¸ ì£¼ì˜ì‚¬í•­:
- ì›¹ìº  ì ‘ê·¼ ê¶Œí•œì„ í—ˆìš©í•´ì£¼ì„¸ìš”
- ë°ì€ ê³³ì—ì„œ ì´¬ì˜í•˜ë©´ ë” ì •í™•í•´ìš”
- ì–¼êµ´ì´ í™”ë©´ì— ì˜ ë³´ì´ë„ë¡ í•´ì£¼ì„¸ìš”
- ì¢…ë£Œí•  ë•ŒëŠ” **ë¹¨ê°„ ë²„íŠ¼ í´ë¦­** ë˜ëŠ” **'q' í‚¤**ë¥¼ ì‚¬ìš©í•˜ì„¸ìš”!

````python
def start_emotion_detection():
    """
    ì‹¤ì‹œê°„ ê°ì • ì¸ì‹ì„ ì‹œì‘í•˜ëŠ” ë©”ì¸ í•¨ìˆ˜
    """
    print("ğŸ¥ ì›¹ìº ì„ ì´ˆê¸°í™”í•˜ëŠ” ì¤‘...")
    
    # ì›¹ìº  ì—°ê²° (0ë²ˆì€ ê¸°ë³¸ ì¹´ë©”ë¼)
    cap = cv2.VideoCapture(0)
    
    if not cap.isOpened():
        print("âŒ ì¹´ë©”ë¼ë¥¼ ì—´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤!")
        print("ğŸ”§ í•´ê²° ë°©ë²•:")
        print("   1. ë‹¤ë¥¸ í”„ë¡œê·¸ë¨ì—ì„œ ì¹´ë©”ë¼ë¥¼ ì‚¬ìš©í•˜ê³  ìˆëŠ”ì§€ í™•ì¸")
        print("   2. ì›¹ìº ì´ ì œëŒ€ë¡œ ì—°ê²°ë˜ì–´ ìˆëŠ”ì§€ í™•ì¸")
        print("   3. ì¹´ë©”ë¼ ê¶Œí•œì´ í—ˆìš©ë˜ì–´ ìˆëŠ”ì§€ í™•ì¸")
        return
    
    print("\nğŸ‰ ê°ì • ì¸ì‹ì´ ì‹œì‘ë˜ì—ˆìŠµë‹ˆë‹¤!")
    print("ğŸ“¹ ì›¹ìº  ì°½ì´ ì—´ë¦½ë‹ˆë‹¤...")
    print("ğŸ­ ë‹¤ì–‘í•œ í‘œì •ì„ ì§€ì–´ë³´ì„¸ìš”!")
    print("ğŸ–±ï¸  ìœˆë„ìš° X ë²„íŠ¼, ë¹¨ê°„ EXIT ë²„íŠ¼, ë˜ëŠ” 'q' í‚¤ë¡œ ì¢…ë£Œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
    print("\nğŸ’¡ íŒ: ë°ì€ ê³³ì—ì„œ ì–¼êµ´ì´ ì˜ ë³´ì´ë„ë¡ í•´ì£¼ì„¸ìš”!")
    
    # ì¢…ë£Œ ë²„íŠ¼ì„ ìœ„í•œ ì „ì—­ ë³€ìˆ˜
    should_exit = False
    
    def mouse_callback(event, x, y, flags, param):
        """
        ë§ˆìš°ìŠ¤ í´ë¦­ ì´ë²¤íŠ¸ë¥¼ ì²˜ë¦¬í•˜ëŠ” í•¨ìˆ˜
        ë¹¨ê°„ ì¢…ë£Œ ë²„íŠ¼ ì˜ì—­ì„ í´ë¦­í•˜ë©´ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤
        """
        nonlocal should_exit
        if event == cv2.EVENT_LBUTTONDOWN:
            # ì¢…ë£Œ ë²„íŠ¼ ì˜ì—­ (ìš°ìƒë‹¨): x ì¢Œí‘œ 520-620, y ì¢Œí‘œ 10-50
            if 520 <= x <= 620 and 10 <= y <= 50:
                print("\nğŸ”´ ì¢…ë£Œ ë²„íŠ¼ì´ í´ë¦­ë˜ì—ˆìŠµë‹ˆë‹¤!")
                should_exit = True
    
    # ìœˆë„ìš° ìƒì„± ë° ë§ˆìš°ìŠ¤ ì½œë°± ì„¤ì •
    window_name = 'ğŸ­ ì‹¤ì‹œê°„ ê°ì • ì¸ì‹ - Xë²„íŠ¼/ë¹¨ê°„ë²„íŠ¼/qí‚¤ë¡œ ì¢…ë£Œ'
    cv2.namedWindow(window_name)
    cv2.setMouseCallback(window_name, mouse_callback)
    
    while True:
        # ì¹´ë©”ë¼ì—ì„œ í•œ í”„ë ˆì„ ì½ê¸°
        ret, frame = cap.read()
        if not ret:
            print("âŒ ì¹´ë©”ë¼ì—ì„œ ì˜ìƒì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            break
        
        # YOLOë¡œ ì‚¬ëŒ ê°ì§€
        yolo_results = yolo_model(frame, verbose=False)
        
        # ê°ì§€ëœ ì‚¬ëŒë“¤ì„ í•˜ë‚˜ì”© ì²˜ë¦¬
        for result in yolo_results:
            boxes = result.boxes
            if boxes is not None:
                for box in boxes:
                    class_id = int(box.cls[0])
                    confidence = float(box.conf[0])
                    
                    # ì‚¬ëŒì´ ê°ì§€ë˜ê³  ì‹ ë¢°ë„ê°€ 50% ì´ìƒì¸ ê²½ìš°
                    if class_id == 0 and confidence > 0.5:
                        x1, y1, x2, y2 = map(int, box.xyxy[0])
                        
                        # ì‚¬ëŒì´ ìˆëŠ” ì˜ì—­ë§Œ ì¶”ì¶œí•´ì„œ ì–¼êµ´ ì°¾ê¸°
                        person_roi = frame[y1:y2, x1:x2]
                        gray_roi = cv2.cvtColor(person_roi, cv2.COLOR_BGR2GRAY)
                        
                        # dlibìœ¼ë¡œ ì–¼êµ´ ê°ì§€
                        faces_dlib = detector(gray_roi)
                        
                        # ê°ì§€ëœ ì–¼êµ´ë§ˆë‹¤ ì²˜ë¦¬
                        for face in faces_dlib:
                            # 68ê°œ ëœë“œë§ˆí¬ í¬ì¸íŠ¸ ì°¾ê¸°
                            landmarks = predictor(gray_roi, face)
                            
                            # ğŸ­ ê°ì • ë¶„ì„í•˜ê¸°
                            emotion, emotion_color, emotion_confidence = analyze_facial_expression(landmarks)
                            
                            # ì–¼êµ´ ì™¸ê³½ì„  ê·¸ë¦¬ê¸° ì¤€ë¹„
                            face_outline = []
                            
                            # í„±ì„  í¬ì¸íŠ¸ë“¤ (0~16ë²ˆ)
                            for i in range(0, 17):
                                x = landmarks.part(i).x + x1
                                y = landmarks.part(i).y + y1
                                face_outline.append([x, y])
                            
                            # ì´ë§ˆ ë¶€ë¶„ ì¶”ê°€ (ëˆˆì¹ ìœ„ìª½)
                            forehead_points = []
                            for i in range(17, 27):  # ëˆˆì¹ í¬ì¸íŠ¸ë“¤
                                x = landmarks.part(i).x + x1
                                y = landmarks.part(i).y + y1 - 20  # ì´ë§ˆ ë†’ì´ ì¡°ì •
                                forehead_points.append([x, y])
                            
                            # ì´ë§ˆ í¬ì¸íŠ¸ë¥¼ ì—­ìˆœìœ¼ë¡œ ì¶”ê°€í•´ì„œ ë‹«íŒ ëª¨ì–‘ ë§Œë“¤ê¸°
                            forehead_points.reverse()
                            face_outline.extend(forehead_points)
                            
                            # numpy ë°°ì—´ë¡œ ë³€í™˜
                            face_outline = np.array(face_outline, np.int32)
                            
                            # ğŸ¨ ê°ì •ì— ë”°ë¥¸ ìƒ‰ìƒìœ¼ë¡œ ì–¼êµ´ ì™¸ê³½ì„  ê·¸ë¦¬ê¸°
                            cv2.polylines(frame, [face_outline], True, emotion_color, 3)
                            
                            # ğŸ“ ê°ì • ì •ë³´ í…ìŠ¤íŠ¸ë¡œ í‘œì‹œ
                            face_x = face.left() + x1
                            face_y = face.top() + y1
                            
                            cv2.putText(frame, f'{emotion} ({emotion_confidence:.2f})', 
                                      (face_x, face_y - 40), 
                                      cv2.FONT_HERSHEY_SIMPLEX, 0.8, emotion_color, 2)
        
        # ğŸ”´ ì¢…ë£Œ ë²„íŠ¼ ê·¸ë¦¬ê¸° (ìš°ìƒë‹¨)
        button_x1, button_y1 = 520, 10  # ë²„íŠ¼ ì‹œì‘ ì¢Œí‘œ
        button_x2, button_y2 = 620, 50  # ë²„íŠ¼ ë ì¢Œí‘œ
        
        # ë¹¨ê°„ìƒ‰ ë°°ê²½ìœ¼ë¡œ ë²„íŠ¼ ê·¸ë¦¬ê¸°
        cv2.rectangle(frame, (button_x1, button_y1), (button_x2, button_y2), (0, 0, 255), -1)
        
        # ë²„íŠ¼ í…Œë‘ë¦¬ (ê²€ì€ìƒ‰)
        cv2.rectangle(frame, (button_x1, button_y1), (button_x2, button_y2), (0, 0, 0), 2)
        
        # "ì¢…ë£Œ" í…ìŠ¤íŠ¸ ì¶”ê°€ (í°ìƒ‰)
        text_x = button_x1 + 15  # í…ìŠ¤íŠ¸ x ì¢Œí‘œ
        text_y = button_y1 + 28  # í…ìŠ¤íŠ¸ y ì¢Œí‘œ
        cv2.putText(frame, 'EXIT', (text_x, text_y), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        
        # ì‚¬ìš©ë²• ì•ˆë‚´ í…ìŠ¤íŠ¸ (í™”ë©´ í•˜ë‹¨)
        help_text = "Click X button, RED button or press 'q' to exit"
        cv2.putText(frame, help_text, (10, frame.shape[0] - 20), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
        
        # í™”ë©´ì— ê²°ê³¼ ë³´ì—¬ì£¼ê¸°
        cv2.imshow(window_name, frame)
        
        # ì¢…ë£Œ ì¡°ê±´ í™•ì¸
        if should_exit:
            print("ğŸ›‘ ì¢…ë£Œ ë²„íŠ¼ìœ¼ë¡œ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break
            
        # ìœˆë„ìš°ê°€ ë‹«í˜”ëŠ”ì§€ í™•ì¸ (X ë²„íŠ¼ í´ë¦­)
        if cv2.getWindowProperty(window_name, cv2.WND_PROP_VISIBLE) < 1:
            print("âŒ ìœˆë„ìš° X ë²„íŠ¼ìœ¼ë¡œ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break
            
        # 'q' í‚¤ë¥¼ ëˆ„ë¥´ë©´ ì¢…ë£Œ (ê¸°ì¡´ ê¸°ëŠ¥ ìœ ì§€)
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            print("âŒ¨ï¸  'q' í‚¤ë¡œ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break
    
    # ì •ë¦¬ ì‘ì—…
    cap.release()           # ì¹´ë©”ë¼ í•´ì œ
    cv2.destroyAllWindows() # ëª¨ë“  ì°½ ë‹«ê¸°
    print("\nğŸ‘‹ ê°ì • ì¸ì‹ì„ ì¢…ë£Œí–ˆìŠµë‹ˆë‹¤. ìˆ˜ê³ í•˜ì…¨ì–´ìš”!")

# ğŸš€ ì‹¤í–‰í•˜ê¸°!
print("ğŸ¬ ì‹¤ì‹œê°„ ê°ì • ì¸ì‹ì„ ì‹œì‘í•©ë‹ˆë‹¤!")
print("ì ê¹, ì¤€ë¹„ ì¤‘...")

try:
    start_emotion_detection()
except KeyboardInterrupt:
    print("\nâ¹ï¸  ì‚¬ìš©ìê°€ ì¤‘ë‹¨í–ˆìŠµë‹ˆë‹¤.")
except Exception as e:
    print(f"\nâŒ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
    print("ğŸ”§ í•´ê²° ë°©ë²•ì„ ìœ„í•´ ìœ„ì˜ ë‹¨ê³„ë“¤ì„ ë‹¤ì‹œ í™•ì¸í•´ë³´ì„¸ìš”.")
````
