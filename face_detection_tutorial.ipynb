# 0단계: pip 설치 확인하기 🔧

**pip**가 뭔가요? 🤔

pip는 **Python Package Installer**의 줄임말이에요. 쉽게 말해서 **파이썬용 앱스토어**라고 생각하면 돼요!

## 📱 스마트폰과 비교해보면:
- **앱스토어/플레이스토어**: 스마트폰에 앱을 다운로드하고 설치
- **pip**: 파이썬에 라이브러리(도구들)를 다운로드하고 설치

## 🎯 왜 필요한가요?
우리가 만들 얼굴 감정 인식 프로그램을 위해서는:
- 카메라를 다루는 도구 📷
- AI 모델을 사용하는 도구 🤖  
- 이미지를 처리하는 도구 🖼️

이런 **전문적인 도구들**이 필요해요. 이 도구들을 **라이브러리**라고 부르고, pip로 설치할 수 있어요!

## ⚠️ 주의사항
- 대부분의 파이썬 설치에는 pip가 포함되어 있어요
- 만약 없다면 아래 코드가 자동으로 설치해줄 거예요
- 걱정하지 마세요! 🙂

자, 그럼 pip가 설치되어 있는지 확인해볼까요?

````python
# 0단계: pip 설치 확인 및 설치 🔧

print("🔍 pip 설치 상태를 확인하는 중...")
print("(pip = 파이썬용 앱스토어라고 생각하세요!)")

import sys
import subprocess
import os

def check_and_install_pip():
    """
    pip가 설치되어 있는지 확인하고, 없으면 설치하는 함수
    
    🤖 이 함수가 하는 일:
    1. pip가 있는지 확인
    2. 없으면 자동으로 설치
    3. 설치 결과를 알려줌
    """
    try:
        # pip 명령어가 있는지 확인 (마치 앱이 설치되어 있는지 확인하는 것처럼)
        result = subprocess.run([sys.executable, '-m', 'pip', '--version'], 
                              capture_output=True, text=True)
        if result.returncode == 0:
            print("✅ pip가 이미 설치되어 있습니다!")
            print(f"   📋 버전 정보: {result.stdout.strip()}")
            print("   🎉 라이브러리 설치 준비 완료!")
            return True
        else:
            print("❌ pip가 설치되어 있지 않습니다.")
    except:
        print("❌ pip가 설치되어 있지 않습니다.")
    
    print("🔧 pip 설치를 자동으로 시도합니다...")
    print("   (잠깐만 기다려주세요!)")
    
    try:
        # ensurepip 모듈을 사용해서 pip 설치
        # (파이썬에 내장된 pip 설치 도구)
        if os.name == 'nt':  # Windows 컴퓨터인 경우
            print("   💻 Windows에서 pip 설치 중...")
        else:  # macOS/Linux 컴퓨터인 경우
            print("   🍎 macOS/Linux에서 pip 설치 중...")
        
        subprocess.run([sys.executable, '-m', 'ensurepip', '--upgrade'], check=True)
        
        print("✅ pip 설치 완료!")
        print("   🎊 이제 필요한 라이브러리들을 설치할 수 있어요!")
        return True
        
    except Exception as e:
        print(f"❌ pip 자동 설치 실패: {e}")
        print("\n🆘 수동 설치 방법을 알려드릴게요:")
        print("\n   🍎 macOS/Linux 사용자:")
        print("   1. 터미널을 열어주세요")
        print("   2. 다음 명령어를 입력하세요:")
        print("      curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py")
        print("   3. 그 다음: python get-pip.py")
        print("\n   💻 Windows 사용자:")
        print("   1. https://bootstrap.pypa.io/get-pip.py 를 다운로드")
        print("   2. 명령 프롬프트에서: python get-pip.py")
        print("\n   🔄 또는 Python을 다시 설치하세요 (pip 포함하여)")
        print("      https://www.python.org 에서 최신 버전 다운로드")
        return False

# 🚀 pip 확인 및 설치 실행!
print("🎬 pip 설치 확인을 시작합니다!")
print("\n💡 pip란? 파이썬 라이브러리를 설치하는 도구예요!")
print("   (스마트폰의 앱스토어 같은 역할을 해요)")

pip_ready = check_and_install_pip()

if pip_ready:
    print("\n🎉 pip 준비 완료!")
    print("✨ 이제 다음 단계에서 멋진 라이브러리들을 설치할 수 있어요!")
    print("📚 OpenCV, YOLO, dlib 같은 AI 도구들을 만나볼 거예요!")
    print("\n➡️  다음 셀로 넘어가세요!")
else:
    print("\n⚠️  pip 설치에 문제가 있어요.")
    print("🔧 위의 수동 설치 방법을 따라한 후 다시 시도해주세요.")
    print("💭 어려우면 선생님이나 친구에게 도움을 요청하세요!")
````

# 1단계: 필요한 라이브러리 설치하기 📦

먼저 우리가 사용할 도구들을 설치해야 합니다. 마치 요리하기 전에 재료를 준비하는 것과 같아요!

아래 코드를 실행하면 다음과 같은 도구들이 설치됩니다:
- **opencv-python**: 카메라와 이미지를 다루는 도구
- **ultralytics**: YOLO AI 모델을 사용하는 도구
- **numpy**: 수학 계산을 도와주는 도구
- **dlib**: 얼굴 분석을 도와주는 도구

⏰ **설치 시간**: 약 2-3분 정도 걸릴 수 있어요. 기다려주세요!

````python
# 필요한 라이브러리들을 설치합니다
# 이 코드를 실행하면 자동으로 설치가 진행됩니다

print("📦 라이브러리 설치를 시작합니다...")
print("☕ 잠깐 커피라도 마시며 기다려주세요!")

import sys
import subprocess
import os

def install_packages():
    """
    필요한 라이브러리들을 설치하는 함수
    macOS의 외부 관리 환경 문제를 해결합니다
    """
    packages = ['opencv-python', 'ultralytics', 'numpy']
    
    for package in packages:
        try:
            print(f"🔧 {package} 설치 중...")
            
            # 첫 번째 시도: 일반 설치
            try:
                result = subprocess.run(
                    [sys.executable, '-m', 'pip', 'install', package], 
                    capture_output=True, text=True, check=True
                )
                print(f"✅ {package} 설치 완료!")
                continue
            except subprocess.CalledProcessError as e:
                if "externally-managed-environment" in str(e.stderr):
                    print(f"   ⚠️  외부 관리 환경 문제 감지. 다른 방법으로 시도합니다...")
                    
                    # 두 번째 시도: --break-system-packages 플래그 사용
                    try:
                        result = subprocess.run(
                            [sys.executable, '-m', 'pip', 'install', '--break-system-packages', package], 
                            capture_output=True, text=True, check=True
                        )
                        print(f"✅ {package} 설치 완료! (시스템 패키지 우회 방식)")
                        continue
                    except subprocess.CalledProcessError as e2:
                        print(f"   ❌ 시스템 패키지 우회 방식도 실패: {e2.stderr}")
                        
                        # 세 번째 시도: --user 플래그 사용
                        try:
                            result = subprocess.run(
                                [sys.executable, '-m', 'pip', 'install', '--user', package], 
                                capture_output=True, text=True, check=True
                            )
                            print(f"✅ {package} 설치 완료! (사용자 설치 방식)")
                            continue
                        except subprocess.CalledProcessError as e3:
                            print(f"   ❌ 사용자 설치 방식도 실패: {e3.stderr}")
                            print(f"\n🔧 {package} 수동 설치 방법:")
                            print(f"   터미널에서 다음 명령어를 실행해보세요:")
                            print(f"   pip3 install --break-system-packages {package}")
                            print(f"   또는: brew install {package}")
                            return False
                else:
                    print(f"❌ {package} 설치 실패: {e}")
                    print(f"   오류 내용: {e.stderr}")
                    return False
                    
        except Exception as e:
            print(f"❌ {package} 설치 중 예상치 못한 오류: {e}")
            return False
    
    return True

# 라이브러리 설치 실행
print("💡 macOS에서 외부 관리 환경 문제가 있을 수 있어요. 여러 방법으로 시도해봅니다!")
success = install_packages()

if success:
    print("\n🎉 모든 라이브러리 설치가 완료되었습니다!")
    print("✨ OpenCV, YOLO, NumPy가 준비되었어요!")
    print("📍 이제 dlib 설치를 진행해봅시다.")
else:
    print("\n⚠️  설치 중 문제가 발생했습니다.")
    print("🔧 다음 해결 방법들을 시도해보세요:")
    print("\n📋 해결 방법 1: 터미널에서 직접 설치")
    print("   pip3 install --break-system-packages opencv-python ultralytics numpy")
    print("\n📋 해결 방법 2: Homebrew로 설치")
    print("   brew install opencv python-numpy")
    print("   pip3 install --break-system-packages ultralytics")
    print("\n📋 해결 방법 3: 가상환경 사용")
    print("   python3 -m venv face_emotion_env")
    print("   source face_emotion_env/bin/activate")
    print("   pip install opencv-python ultralytics numpy")
    print("\n💭 어려우면 선생님이나 친구에게 도움을 요청하세요!")
````

## dlib 설치하기 🔧

**dlib**은 조금 특별한 라이브러리예요. 설치가 어려울 수 있으니 여러 방법을 시도해볼게요!

💡 **팁**: 만약 아래 코드에서 오류가 난다면, 다음 방법들을 시도해보세요:
1. 먼저 아래 코드를 실행해보세요
2. 오류가 나면 터미널에서 `conda install dlib` 시도
3. 그래도 안 되면 `pip install dlib-binary` 시도

````python
# dlib 설치를 시도합니다
print("🔧 dlib 설치를 시도합니다...")
print("💡 dlib은 조금 특별한 라이브러리라서 설치에 시간이 걸릴 수 있어요!")

import sys
import subprocess

def install_dlib():
    """
    dlib 라이브러리를 설치하는 함수
    여러 방법을 시도해서 설치 성공률을 높입니다
    macOS의 외부 관리 환경 문제도 해결합니다
    """
    
    # 방법 1: 일반 dlib 설치 시도
    try:
        print("\n🎯 방법 1: 일반 dlib 설치 시도...")
        result = subprocess.run(
            [sys.executable, '-m', 'pip', 'install', 'dlib'], 
            capture_output=True, text=True, timeout=300  # 5분 타임아웃
        )
        if result.returncode == 0:
            print("✅ dlib 설치 성공!")
            return True
        else:
            if "externally-managed-environment" in str(result.stderr):
                print("   ⚠️  외부 관리 환경 문제 감지. 다른 방법으로 시도합니다...")
                
                # --break-system-packages로 재시도
                try:
                    result = subprocess.run(
                        [sys.executable, '-m', 'pip', 'install', '--break-system-packages', 'dlib'], 
                        capture_output=True, text=True, timeout=300
                    )
                    if result.returncode == 0:
                        print("✅ dlib 설치 성공! (시스템 패키지 우회 방식)")
                        return True
                except subprocess.TimeoutExpired:
                    print("   ⏰ 시스템 패키지 우회 방식 시간 초과")
                except Exception as e:
                    print(f"   ❌ 시스템 패키지 우회 방식 실패: {e}")
            else:
                print(f"❌ 일반 dlib 설치 실패: {result.stderr}")
    except subprocess.TimeoutExpired:
        print("⏰ dlib 설치 시간 초과")
    except Exception as e:
        print(f"❌ dlib 설치 중 오류: {e}")
    
    # 방법 2: dlib-binary 설치 시도
    try:
        print("\n🎯 방법 2: dlib-binary 설치 시도...")
        result = subprocess.run(
            [sys.executable, '-m', 'pip', 'install', 'dlib-binary'], 
            capture_output=True, text=True, check=True
        )
        print("✅ dlib-binary 설치 성공!")
        return True
    except subprocess.CalledProcessError as e:
        if "externally-managed-environment" in str(e.stderr):
            print("   ⚠️  외부 관리 환경 문제 감지. 우회 방식으로 시도합니다...")
            try:
                result = subprocess.run(
                    [sys.executable, '-m', 'pip', 'install', '--break-system-packages', 'dlib-binary'], 
                    capture_output=True, text=True, check=True
                )
                print("✅ dlib-binary 설치 성공! (시스템 패키지 우회 방식)")
                return True
            except subprocess.CalledProcessError as e2:
                print(f"❌ dlib-binary 우회 설치 실패: {e2.stderr}")
        else:
            print(f"❌ dlib-binary 설치 실패: {e.stderr}")
    except Exception as e:
        print(f"❌ dlib-binary 설치 중 오류: {e}")
    
    return False

# dlib 설치 실행
success = install_dlib()

if success:
    print("\n🎉 dlib 설치 완료!")
    print("✨ 이제 모든 라이브러리가 준비되었습니다!")
    print("📍 다음 단계로 넘어가세요!")
else:
    print("\n⚠️  dlib 설치에 문제가 있습니다.")
    print("🔧 다음 방법들을 시도해보세요:")
    print("\n📋 해결 방법 1: 터미널에서 직접 설치")
    print("   pip3 install --break-system-packages dlib-binary")
    print("\n📋 해결 방법 2: Homebrew로 설치")
    print("   brew install dlib")
    print("\n📋 해결 방법 3: 가상환경에서 설치")
    print("   python3 -m venv face_emotion_env")
    print("   source face_emotion_env/bin/activate")
    print("   pip install dlib-binary")
    print("\n💭 dlib 없이도 기본 얼굴 감지는 가능하니 일단 다음 단계로 넘어가도 됩니다!")
````

# 3단계: AI 모델 다운로드하기 🤖

이제 얼굴 분석을 위한 특별한 AI 모델을 다운로드해야 합니다.

## 무엇을 다운로드하나요?
**shape_predictor_68_face_landmarks.dat** 파일을 다운로드합니다.
- 이 파일은 얼굴에서 눈, 코, 입, 턱선 등 68개의 중요한 점들을 찾아주는 AI 모델이에요

# 4단계: 감정 분석 함수 만들기 😊😢😠😲

이제 가장 재미있는 부분입니다! 얼굴 표정을 보고 감정을 알아내는 함수를 만들어봅시다.

## 어떻게 감정을 알아낼까요?

우리 함수는 얼굴의 **68개 중요 포인트**를 분석합니다:

### 1. 입 모양 분석 👄
- **웃음**: 입꼬리가 올라가고 입이 옆으로 길어짐
- **슬픔**: 입꼬리가 아래로 내려감

### 2. 눈 크기 분석 👀
- **놀람**: 눈이 평소보다 크게 뜨임
- **평상시**: 적당한 크기

### 3. 눈썹 위치 분석 🤨
- **화남**: 눈썹이 내려와서 찡그린 모양
- **놀람**: 눈썹이 올라감

### 감지할 수 있는 감정들:
- 😊 **Happy** (행복) - 초록색으로 표시
- 😢 **Sad** (슬픔) - 빨간색으로 표시  
- 😠 **Angry** (화남) - 진한 빨간색으로 표시
- 😲 **Surprised** (놀람) - 노란색으로 표시
- 😐 **Neutral** (평상시) - 흰색으로 표시

````python
def analyze_facial_expression(landmarks):
    """
    얼굴 랜드마크를 분석하여 감정을 판단하는 함수
    
    입력: landmarks (얼굴의 68개 중요 포인트)
    출력: (감정이름, 색상, 신뢰도)
    """
    
    # 🗣️ 입 관련 포인트들을 가져옵니다
    mouth_left = landmarks.part(48)    # 입 왼쪽 끝
    mouth_right = landmarks.part(54)   # 입 오른쪽 끝
    mouth_top = landmarks.part(51)     # 입 위쪽
    mouth_bottom = landmarks.part(57)  # 입 아래쪽
    
    # 👀 눈 관련 포인트들을 가져옵니다
    left_eye_top = landmarks.part(37)     # 왼쪽 눈 위
    left_eye_bottom = landmarks.part(41)  # 왼쪽 눈 아래
    right_eye_top = landmarks.part(43)    # 오른쪽 눈 위
    right_eye_bottom = landmarks.part(47) # 오른쪽 눈 아래
    
    # 🤨 눈썹 관련 포인트들을 분석합니다
    left_eyebrow = np.mean([landmarks.part(i).y for i in range(17, 22)])   # 왼쪽 눈썹 평균 높이
    right_eyebrow = np.mean([landmarks.part(i).y for i in range(22, 27)])  # 오른쪽 눈썹 평균 높이
    
    # 📊 특징들을 계산합니다
    
    # 1️⃣ 입의 가로세로 비율 (웃을 때는 입이 옆으로 길어져요)
    mouth_width = abs(mouth_right.x - mouth_left.x)      # 입 너비
    mouth_height = abs(mouth_top.y - mouth_bottom.y)     # 입 높이
    mouth_ratio = mouth_height / mouth_width if mouth_width > 0 else 0
    
    # 2️⃣ 입꼬리 기울기 (웃으면 올라가고, 슬프면 내려가요)
    mouth_curve = (mouth_left.y + mouth_right.y) / 2 - mouth_top.y
    
    # 3️⃣ 눈의 크기 (놀라면 눈이 커져요)
    left_eye_height = abs(left_eye_top.y - left_eye_bottom.y)
    right_eye_height = abs(right_eye_top.y - right_eye_bottom.y)
    avg_eye_height = (left_eye_height + right_eye_height) / 2
    
    # 4️⃣ 눈썹과 눈 사이의 거리 (화나면 눈썹이 내려와요)
    avg_eyebrow_height = (left_eyebrow + right_eyebrow) / 2
    eye_center = (left_eye_top.y + right_eye_top.y) / 2
    eyebrow_distance = eye_center - avg_eyebrow_height
    
    # 🎯 감정을 판단합니다!
    emotion = "Neutral"         # 기본값: 평상시
    color = (255, 255, 255)     # 기본 색상: 흰색
    confidence = 0.5            # 기본 신뢰도: 50%
    
    # 조건에 따라 감정을 판별합니다
    if mouth_curve < -5 and mouth_ratio < 0.5:  # 😊 웃음: 입꼬리 올라가고 입이 넓어짐
        emotion = "Happy"
        color = (0, 255, 0)  # 초록색
        confidence = min(1.0, abs(mouth_curve) / 10)
        
    elif mouth_curve > 7:  # 😢 슬픔: 입꼬리 내려감
        emotion = "Sad"
        color = (255, 0, 0)  # 빨간색
        confidence = min(1.0, mouth_curve / 10)
        
    elif avg_eye_height > 8 and eyebrow_distance > 15:  # 😲 놀람: 눈 크게 뜨고 눈썹 올라감
        emotion = "Surprised"
        color = (0, 255, 255)  # 노란색
        confidence = min(1.0, avg_eye_height / 10)
        
    elif eyebrow_distance < 10:  # 😠 화남: 눈썹이 내려와서 찡그림
        emotion = "Angry"
        color = (0, 0, 255)  # 진한 빨간색
        confidence = min(1.0, (15 - eyebrow_distance) / 10)
    
    return emotion, color, confidence

print("✅ 감정 분석 함수가 준비되었습니다!")
print("이제 AI 모델들을 불러와봅시다.")
````

# 6단계: 실시간 감정 인식 실행하기! 🎥✨

드디어 모든 준비가 끝났습니다! 이제 웹캠을 켜서 실시간으로 감정을 인식해봅시다.

## 🎮 사용법:
1. 아래 코드를 실행하면 웹캠 창이 열립니다
2. 카메라 앞에 얼굴을 보여주세요
3. 다양한 표정을 지어보세요! (웃기, 슬퍼하기, 화내기, 놀라기)
4. 얼굴 주변에 색깔이 있는 외곽선이 나타나고 감정이 표시됩니다
5. **우상단의 빨간 'EXIT' 버튼을 클릭**하거나 **'q' 키를 누르면 종료**됩니다

## 🌈 색깔 의미:
- **초록색** 😊 = Happy (행복)
- **빨간색** 😢 = Sad (슬픔)
- **진한 빨간색** 😠 = Angry (화남)
- **노란색** 😲 = Surprised (놀람)
- **흰색** 😐 = Neutral (평상시)

## 🖱️ 종료 방법:
- **방법 1**: 윈도우 우상단의 **X 버튼** 클릭 (기본 윈도우 종료)
- **방법 2**: 화면 내 빨간 **EXIT** 버튼을 마우스로 클릭
- **방법 3**: 키보드에서 **'q'** 키 누르기

## ⚠️ 주의사항:
- 웹캠 접근 권한을 허용해주세요
- 밝은 곳에서 촬영하면 더 정확해요
- 얼굴이 화면에 잘 보이도록 해주세요
- 종료할 때는 **빨간 버튼 클릭** 또는 **'q' 키**를 사용하세요!

````python
def start_emotion_detection():
    """
    실시간 감정 인식을 시작하는 메인 함수
    """
    print("🎥 웹캠을 초기화하는 중...")
    
    # 웹캠 연결 (0번은 기본 카메라)
    cap = cv2.VideoCapture(0)
    
    if not cap.isOpened():
        print("❌ 카메라를 열 수 없습니다!")
        print("🔧 해결 방법:")
        print("   1. 다른 프로그램에서 카메라를 사용하고 있는지 확인")
        print("   2. 웹캠이 제대로 연결되어 있는지 확인")
        print("   3. 카메라 권한이 허용되어 있는지 확인")
        return
    
    print("\n🎉 감정 인식이 시작되었습니다!")
    print("📹 웹캠 창이 열립니다...")
    print("🎭 다양한 표정을 지어보세요!")
    print("🖱️  윈도우 X 버튼, 빨간 EXIT 버튼, 또는 'q' 키로 종료할 수 있습니다.")
    print("\n💡 팁: 밝은 곳에서 얼굴이 잘 보이도록 해주세요!")
    
    # 종료 버튼을 위한 전역 변수
    should_exit = False
    
    def mouse_callback(event, x, y, flags, param):
        """
        마우스 클릭 이벤트를 처리하는 함수
        빨간 종료 버튼 영역을 클릭하면 프로그램을 종료합니다
        """
        nonlocal should_exit
        if event == cv2.EVENT_LBUTTONDOWN:
            # 종료 버튼 영역 (우상단): x 좌표 520-620, y 좌표 10-50
            if 520 <= x <= 620 and 10 <= y <= 50:
                print("\n🔴 종료 버튼이 클릭되었습니다!")
                should_exit = True
    
    # 윈도우 생성 및 마우스 콜백 설정
    window_name = '🎭 실시간 감정 인식 - X버튼/빨간버튼/q키로 종료'
    cv2.namedWindow(window_name)
    cv2.setMouseCallback(window_name, mouse_callback)
    
    while True:
        # 카메라에서 한 프레임 읽기
        ret, frame = cap.read()
        if not ret:
            print("❌ 카메라에서 영상을 읽을 수 없습니다.")
            break
        
        # YOLO로 사람 감지
        yolo_results = yolo_model(frame, verbose=False)
        
        # 감지된 사람들을 하나씩 처리
        for result in yolo_results:
            boxes = result.boxes
            if boxes is not None:
                for box in boxes:
                    class_id = int(box.cls[0])
                    confidence = float(box.conf[0])
                    
                    # 사람이 감지되고 신뢰도가 50% 이상인 경우
                    if class_id == 0 and confidence > 0.5:
                        x1, y1, x2, y2 = map(int, box.xyxy[0])
                        
                        # 사람이 있는 영역만 추출해서 얼굴 찾기
                        person_roi = frame[y1:y2, x1:x2]
                        gray_roi = cv2.cvtColor(person_roi, cv2.COLOR_BGR2GRAY)
                        
                        # dlib으로 얼굴 감지
                        faces_dlib = detector(gray_roi)
                        
                        # 감지된 얼굴마다 처리
                        for face in faces_dlib:
                            # 68개 랜드마크 포인트 찾기
                            landmarks = predictor(gray_roi, face)
                            
                            # 🎭 감정 분석하기
                            emotion, emotion_color, emotion_confidence = analyze_facial_expression(landmarks)
                            
                            # 얼굴 외곽선 그리기 준비
                            face_outline = []
                            
                            # 턱선 포인트들 (0~16번)
                            for i in range(0, 17):
                                x = landmarks.part(i).x + x1
                                y = landmarks.part(i).y + y1
                                face_outline.append([x, y])
                            
                            # 이마 부분 추가 (눈썹 위쪽)
                            forehead_points = []
                            for i in range(17, 27):  # 눈썹 포인트들
                                x = landmarks.part(i).x + x1
                                y = landmarks.part(i).y + y1 - 20  # 이마 높이 조정
                                forehead_points.append([x, y])
                            
                            # 이마 포인트를 역순으로 추가해서 닫힌 모양 만들기
                            forehead_points.reverse()
                            face_outline.extend(forehead_points)
                            
                            # numpy 배열로 변환
                            face_outline = np.array(face_outline, np.int32)
                            
                            # 🎨 감정에 따른 색상으로 얼굴 외곽선 그리기
                            cv2.polylines(frame, [face_outline], True, emotion_color, 3)
                            
                            # 📝 감정 정보 텍스트로 표시
                            face_x = face.left() + x1
                            face_y = face.top() + y1
                            
                            cv2.putText(frame, f'{emotion} ({emotion_confidence:.2f})', 
                                      (face_x, face_y - 40), 
                                      cv2.FONT_HERSHEY_SIMPLEX, 0.8, emotion_color, 2)
        
        # 🔴 종료 버튼 그리기 (우상단)
        button_x1, button_y1 = 520, 10  # 버튼 시작 좌표
        button_x2, button_y2 = 620, 50  # 버튼 끝 좌표
        
        # 빨간색 배경으로 버튼 그리기
        cv2.rectangle(frame, (button_x1, button_y1), (button_x2, button_y2), (0, 0, 255), -1)
        
        # 버튼 테두리 (검은색)
        cv2.rectangle(frame, (button_x1, button_y1), (button_x2, button_y2), (0, 0, 0), 2)
        
        # "종료" 텍스트 추가 (흰색)
        text_x = button_x1 + 15  # 텍스트 x 좌표
        text_y = button_y1 + 28  # 텍스트 y 좌표
        cv2.putText(frame, 'EXIT', (text_x, text_y), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        
        # 사용법 안내 텍스트 (화면 하단)
        help_text = "Click X button, RED button or press 'q' to exit"
        cv2.putText(frame, help_text, (10, frame.shape[0] - 20), 
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
        
        # 화면에 결과 보여주기
        cv2.imshow(window_name, frame)
        
        # 종료 조건 확인
        if should_exit:
            print("🛑 종료 버튼으로 프로그램을 종료합니다.")
            break
            
        # 윈도우가 닫혔는지 확인 (X 버튼 클릭)
        if cv2.getWindowProperty(window_name, cv2.WND_PROP_VISIBLE) < 1:
            print("❌ 윈도우 X 버튼으로 프로그램을 종료합니다.")
            break
            
        # 'q' 키를 누르면 종료 (기존 기능 유지)
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            print("⌨️  'q' 키로 프로그램을 종료합니다.")
            break
    
    # 정리 작업
    cap.release()           # 카메라 해제
    cv2.destroyAllWindows() # 모든 창 닫기
    print("\n👋 감정 인식을 종료했습니다. 수고하셨어요!")

# 🚀 실행하기!
print("🎬 실시간 감정 인식을 시작합니다!")
print("잠깐, 준비 중...")

try:
    start_emotion_detection()
except KeyboardInterrupt:
    print("\n⏹️  사용자가 중단했습니다.")
except Exception as e:
    print(f"\n❌ 오류가 발생했습니다: {e}")
    print("🔧 해결 방법을 위해 위의 단계들을 다시 확인해보세요.")
````
